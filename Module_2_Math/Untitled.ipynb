```python
import scipy.stats as st

def z_table():
    alpha_levels = [.001, .01, .05, .1]
    z_values = [x / 10.0 for x in range(0, 61)]
    print("z", end='\t')
    for alpha in alpha_levels:
        print(f"{alpha:.3f}", end='\t')
    print()
    for z in z_values:
        print(f"{z:.1f}", end='\t')
        for alpha in alpha_levels:
            p = st.norm.sf(abs(z)) * 2
            if p < alpha:
                print("**", end='\t')
            else:
                print("  ", end='\t')
        print()

z_table()
```
