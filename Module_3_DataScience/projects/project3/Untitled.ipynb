```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# load the data
df = pd.read_csv('C:/Users/realragnvaldr/Dropbox/Integrify/homework/Module_3_DataScience/projects/project6/data/train.csv')

# compute the correlation matrix
corr = df.corr()

# generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})

plt.show()

# create a mask to remove duplicate / self correlations
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)

# apply the mask and stack to get a series of correlations
corr_pairs = corr.where(np.tril(np.ones(corr.shape)).astype(np.bool)).stack()

# print correlations sorted by absolute value
print("\nCorrelations sorted from low to high:")
for pair, corr_value in corr_pairs.abs().sort_values().items():
    print(f"{pair}: r = {corr_value:.3f}")
```
