```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the normal distribution
mean = -0.1
std_dev = 0.1

# Generate three sets of random numbers from the normal distribution
data1 = np.random.normal(mean, std_dev, 28)
data2 = np.random.normal(mean, std_dev, 28)
data3 = np.random.normal(mean, std_dev, 28)

# Create the histograms
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)  # 1 row, 3 columns, subplot 1
plt.hist(data1, bins='auto', color='skyblue', edgecolor='black')
plt.title("Histogram 1")

plt.subplot(1, 3, 2)  # 1 row, 3 columns, subplot 2
plt.hist(data2, bins='auto', color='skyblue', edgecolor='black')
plt.title("Histogram 2")

plt.subplot(1, 3, 3)  # 1 row, 3 columns, subplot 3
plt.hist(data3, bins='auto', color='skyblue', edgecolor='black')
plt.title("Histogram 3")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the normal distributions
mean = 0.1
std_dev1 = 0.1
std_dev2 = 1

# Generate two sets of random numbers from the normal distributions
data1 = np.random.normal(mean, std_dev1, 28)
data2 = np.random.normal(mean, std_dev2, 28)

# Create the histograms in the same plot
plt.figure(figsize=(10, 6))

# Histogram 1
plt.hist(data1, bins='auto', color='skyblue', edgecolor='black', alpha=0.7, label='sd=0.1')

# Histogram 2
plt.hist(data2, bins='auto', color='orange', edgecolor='black', alpha=0.7, label='sd=1')

plt.title("Histograms of Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters for the normal distributions
mean = -0.1
std_dev1 = 1
std_dev2 = .1

# Generate two sets of random numbers from the normal distributions
data1 = np.random.normal(mean, std_dev1, 28)
data2 = np.random.normal(mean, std_dev2, 28)

# Create the histograms in the same plot
plt.figure(figsize=(10, 6))

# Set a common number of bins
bins = 10

# Histogram 1
plt.hist(data1, bins=bins, color='skyblue', edgecolor='black', alpha=0.7, label='sd=0.1')

# Histogram 2
plt.hist(data2, bins=bins, color='orange', edgecolor='black', alpha=0.7, label='sd=1')

plt.title("Histograms of Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()

plt.tight_layout()
plt.show()
```
