### About dataset
Source: https://www.kaggle.com/datasets/alexteboul/diabetes-health-indicators-dataset

*This is a clean dataset of 253,680 survey responses to the CDC's BRFSS2015. The target variable Diabetes_012 has 3 classes. 0 is for no diabetes or only during pregnancy, 1 is for prediabetes, and 2 is for diabetes. There is class imbalance in this dataset. This dataset has 21 feature variables*

The data file of the following columns:

**Diabetes_binary**
0 = no diabetes 1 = diabetes

**HighBP**
0 = no high BP 1 = high BP

**HighChol**
0 = no high cholesterol 1 = high cholesterol

**CholCheck**
0 = no cholesterol check in 5 years 1 = yes cholesterol check in 5 years

**BMI**
Body Mass Index

**Smoker**
Have you smoked at least 100 cigarettes in your entire life? [Note: 5 packs = 100 cigarettes] 0 = no 1 = yes

**Stroke**
(Ever told) you had a stroke. 0 = no 1 = yes

**HeartDiseaseorAttack**
coronary heart disease (CHD) or myocardial infarction (MI) 0 = no 1 = yes

**PhysActivity**
physical activity in past 30 days - not including job 0 = no 1 = yes

**Fruits**
Consume Fruit 1 or more times per day 0 = no 1 = yes 

**Veggies**
Consume Vegetables 1 or more times per day 0 = no 1 = yes

**HvyAlcoholConsump**
Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week) 0 = no 1 = yes

**AnyHealthcare**
Have any kind of health care coverage, including health insurance, prepaid plans such as HMO, etc. 0 = no 1 = yes

**NoDocbcCost**
Was there a time in the past 12 months when you needed to see a doctor but could not because of cost? 0 = no 1 = yes

**GenHlth**
Would you say that in general your health is: scale 1-5 1 = excellent 2 = very good 3 = good 4 = fair 5 = poor

**MentHlth**
Now thinking about your mental health, which includes stress, depression, and problems with emotions, for how many days during the past 30 days was your mental health not good? scale 1-30 days

**PhysHlth**
Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30 days was your physical health not good? scale 1-30 days

**DiffWalk**
Do you have serious difficulty walking or climbing stairs? 0 = no 1 = yes

**Sex**
0 = female 1 = male

**Age**
13-level age category (AGEG5YR see codebook) 1 = 18-24 9 = 60-64 13 = 80 or older

**Education**
Education level (EDUCA see codebook) scale 1-6 1 = Never attended school or only kindergarten 2 = Grades 1 through 8 (Elementary) 3 = Grades 9 through 11 (Some high school) 4 = Grade 12 or GED (High school graduate) 5 = College 1 year to 3 years (Some college or technical school) 6 = College 4 years or more (College graduate)

**Income**
Income scale (INCOME2 see codebook) scale 1-8. 1 = less than 10k USD; 5 = less than 35k USD; 8 = 75k+ USD

## Imports and settings

```python
# general imports
import os
import pandas as pd
import sys

# import my personal library
sys.path.insert(0, '../../../ronaldlib/')
import ronaldlib.utils as rutils
from ronaldlib.aida import AIDataAnalyzer

# settings
pd.options.display.float_format = "{:.2f}".format
```

## Data loading

```python
# load the data
df = rutils.load_data(data_file, resave_as_pickle=True)
```

## Initialize the AI Analysis class

```python
ai_analyzer = AIDataAnalyzer(df)
ai_analyzer.set_openai_api_key(filename='c:/my_openai_api_key.txt')
ai_analyzer.set_column_descriptions({ 
    'Diabetes_binary'      : '0 = no diabetes 1 = diabetes (binary)',
    'HighBP'               : '0 = no high BP 1 = high BP (binary)',
    'HighChol'             : '0 = no high cholesterol 1 = high cholesterol (binary)',
    'CholCheck'            : '0 = no cholesterol check in 5 years 1 = yes cholesterol check in 5 years (binary)',
    'BMI'                  : 'Body Mass Index (continuous)',
    'Smoker'               : 'Smoked at least 100 cigarettes in their life, 0 = no 1 = yes (binary)',
    'Stroke'               : '(Ever told) you had a stroke. 0 = no 1 = yes (binary)',
    'HeartDiseaseorAttack' : 'coronary heart disease (CHD) or myocardial infarction (MI) 0 = no 1 = yes (binary)',
    'PhysActivity'         : 'physical activity in past 30 days - not including job 0 = no 1 = yes (binary)',
    'Fruits'               : 'Consume Fruit 1 or more times per day 0 = no 1 = yes (binary)',
    'Veggies'              : 'Consume Vegetables 1 or more times per day 0 = no 1 = yes (binary)',
    'HvyAlcoholConsump'    : 'Heavy drinkers, 0 = no 1 = yes (binary)',
    'AnyHealthcare'        : 'Have any kind of health care coverage, including health insurance, prepaid plans such as HMO, etc. 0 = no 1 = yes (binary)',
    'NoDocbcCost'          : 'Was there a time in the past 12 months when you needed to see a doctor but could not because of cost? 0 = no 1 = yes (binary)',
    'GenHlth'              : 'Would you say that in general your health is: scale 1-5 1 = excellent 2 = very good 3 = good 4 = fair 5 = poor (ordinal)',
    'MentHlth'             : 'Number of days of bad mental health in the past 30 days (0 to 30; continuous)',
    'PhysHlth'             : 'Number of days of bad physical health in the past 30 days (0 to 30; continuous)',
    'DiffWalk'             : 'Serious difficulty walking or climbing stairs; 0 = no 1 = yes (binary)',
    'Sex'                  : '0 = female 1 = male (binary)',
    'Age'                  : '13-level age category, 1 = 18-24 9 = 60-64 13 = 80 or older (ordinal)',
    'Education'            : 'Education level, 1 = Never attended school or only kindergarten 2 = Elementary 3 = Some high school 4 = High school graduate 5 = Some college or technical school 6 = College graduate (ordinal)',
    'Income'               : 'Income scale 1-8. 1 = less than 10k USD; 5 = less than 35k USD; 8 = 75k+ USD (nomimal)'
})

# give the analyzer some general information (we'll add this to the gpt prompts)
ai_analyzer.add_comment("The main goal of this project is to understand the factors behind diabetes")
ai_analyzer.add_comment("When you produce plots in a loop, organize them in 1 figure with 3 panels per row")
ai_analyzer.add_comment("Don't forget to label the x- and y-axis in the figure and provide a legend")
ai_analyzer.add_comment("After creating a figure, remove all empty subplots. And make sure there is enough space between the subplots")
```

## Inspecting the data columns: histograms

```python
ai_analyzer.reset_query()
ai_analyzer.build_query("Create a histogram for each data column")
ai_analyzer.execute_query()
```

## Inspecting the data columns: diabetes prevalence for each feature value

```python
ai_analyzer.reset_query()
ai_analyzer.build_query("For this analysis, Diabetes_binary is the target value")
ai_analyzer.build_query("Make bar plots that show the target values as a function of eacht feature")
ai_analyzer.build_query("Display the p value from a chi square test in each plot title")
ai_analyzer.execute_query(run_code=True, verbose_level=0)
```

## Hypothesis testing: chi square tests 

```python
ai_analyzer.reset_query()
ai_analyzer.build_query("In this analysis, Diabetes_binary is the target")
ai_analyzer.build_query("For each feature column with a most 2 unique values, create a contingency table crossed with the target")
ai_analyzer.build_query("Show the table with a nice layout, using IPython display and markdown")
ai_analyzer.build_query("Also show the p value of a chi square test. Leave some white space between the tables")
ai_analyzer.execute_query(run_code=True, verbose_level=0)
```

## Inspecting the data columns: correlation matrix

```python
ai_analyzer.reset_query()
ai_analyzer.build_query("In this analysis, Diabetes_binary is the target")
ai_analyzer.build_query("Compute the Spearman correlation between each feature and the target")
ai_analyzer.build_query("Thereafter, make a correlation heatmap (1 figure, 1 panel)")
ai_analyzer.build_query("Including only features that correlate with at least .15 with the target")
ai_analyzer.execute_query(run_code=True, verbose_level=0)
```

## Inspecting outliers: boxplots

```python
ai_analyzer.reset_query()
ai_analyzer.build_query("Produce a box plot for each continuous feature that has at least 10 unique values")
ai_analyzer.execute_query(run_code=True, verbose_level=0)
```

## Classifcation analysis

```python
ai_analyzer.reset_query()
ai_analyzer.build_query("Train a logistic regression classifier to predict Diabetes_binary")
ai_analyzer.build_query("For all non-binary columns, use StandardScalar for feature scaling")
ai_analyzer.build_query("Use RandomizedSearchCV with 20 iterations and verbose level 1")
ai_analyzer.build_query("Show a performance report in text form")
ai_analyzer.build_query("Show a plot of the confusion matrix")
ai_analyzer.build_query("Show a plot of the feature importance")
ai_analyzer.execute_query(run_code=True, verbose_level=0)
ai_analyzer.reset_query()
```
