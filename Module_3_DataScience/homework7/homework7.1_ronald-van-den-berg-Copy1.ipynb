Suppose we want to measure the water flow using three sensors (S1, S2, S3) fixed in different places. Overalls we made 5 configurations where the data are recorded in  **First set of data.xslsx**:
* Explore the dataset in terms of:
  * Types of variables (categorical or numerical)
  * Statistics (min, mean, max, and std)
  * Errors (if any)
  * Outliers (BE CAREFUL)
  * Are there any errors included in the outlier set?


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# read the data
df = pd.read_csv('group3.csv', sep = ';')
print('Number of rows    = ' + str(df.shape[0]))
print('Number of columns = ' + str(df.shape[1]))

# compute central tendency and spread measures
for col_name, col_data in df.items():
    print(f'\n** Sensor {col_name} **')
    print(f'min, max     = {col_data.min():.1f}, {col_data.max():.1f}')
    print(f'mean, median = {col_data.mean():.1f}, {col_data.median():.1f}')
    print(f'std, var     = {col_data.std():.1f}, {col_data.var():.1f}')
    print(f'skew         = {col_data.skew():.1f}')
    
# create a figure with two subplots
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# boxplot
df.plot(kind='box', ax=axs[0])

# histograms
for col_name in df.columns:
    axs[1].hist(df[col_name], bins=20, alpha=0.5, label=col_name)
    axs[1].legend()
```

# Observations
There are no clear indications of measurement errors. 

Even though each sensor has a few data points that are labeled as outlier (defined as points that are more than 1.5 IQR away from the median), none of them seem so extreme to cause any suspicion.

There is variation between sensor measurement in both their means and spread. Two potential explanations:
1. the water flow differs across the locations where the sensors were placed
2. the sensors differ in measurement precision and bias (this would indicate that the quality of the sensors isn't great)

Without further information, it is impossible to make strong statements about the nature of the measurement differences.
